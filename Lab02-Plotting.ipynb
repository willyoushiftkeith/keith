So, let's load the library:

```python
import matplotlib.pyplot as plt
```

Let's install the numpy library too, so we can use scientific funcions:

```python
import numpy as np
```

We want our plots to appear on this notebook page (not in another window on our computer), so we need to use this command:

```python
%matplotlib inline
```

Now, let's plot something!

How about a simple sine curve?

For this, we'll need a whole load of x values. We'll use a numpy function called linspace that creates evenly spaced values. Use linspace like this:

`np.linspace(start value, end value, number of values)`

```python
x = np.linspace(0, 10, 21)
print(x)
```

Now get the sine of those values:

```python
y = np.sin(x)
print(y)
```

And plot:

```python
plt.plot(x, y, linestyle='-', color='green', linewidth=5)
```

Let's include a Title, and labels on the x-axis and y-axis, and change the colour, linewidth and linestyle:

http://matplotlib.org/examples/lines_bars_and_markers/line_styles_reference.html

```python
plt.title('An Outstanding Plot')
plt.xlabel('My brilliant values for x')
plt.ylabel('The splendid y values')
plt.plot(x, y, linestyle='--', color='brown', linewidth=2)
```

What if we want to plot specific data values? We do this with markers:

http://matplotlib.org/examples/lines_bars_and_markers/marker_reference.html

We can show grid lines on our plot too, using:

'plt.grid(True)'

```python
plt.title('An Outstanding Plot')
plt.xlabel('My brilliant values for x')
plt.ylabel('The splendid y values')
plt.plot(x, y, linestyle=':', color='purple', linewidth=2)
plt.plot(x[5], y[5], marker='o', markersize=10, color='red')
plt.plot(x[11], y[11], color='pink', marker='*', markersize=20)
plt.grid(True)
```
